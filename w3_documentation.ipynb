## Scenario
A student has marks for three quizzes. The program enters the marks, calculates the average, displays the result, decides pass or fail, and repeats for another student if needed.

## 1. Identify the Components
### Inputs
- Three quiz marks.
- A yes/no answer to continue.

### Process
- Read the three marks.
- Find the average.
- Check if the average is 50 or above.
- Repeat for another student if the user wants.

### Output
- Average mark.
- PASS or FAIL.

## 2. Design the Algorithm
### Pseudocode
START
REPEAT
1. Enter Quiz 1, Quiz 2, and Quiz 3 marks.
2. Calculate average = (Quiz 1 + Quiz 2 + Quiz 3) / 3.
3. Display the average.
4. If average >= 50, display PASS; otherwise display FAIL.
5. Ask if another student's marks should be entered.
UNTIL the user types No
END

### Algorithm Structure
#### Sequence
- The program gets the marks, calculates the average, and shows the result in order.

#### Selection
- An if statement is used to decide PASS or FAIL.

#### Iteration
- A while loop repeats the program for another student.

## 3. Final Python Program
```python
while True:
    # Enter three quiz marks.
    quiz1 = float(input("Quiz 1: "))
    quiz2 = float(input("Quiz 2: "))
    quiz3 = float(input("Quiz 3: "))

    # Calculate the average.
    average = (quiz1 + quiz2 + quiz3) / 3
    print("Average mark:", round(average, 2))

    # Check pass or fail.
    if average >= 50:
        print("Result: PASS")
    else:
        print("Result: FAIL")

    # Ask if another student should be entered.
    again = input("Enter another student's marks? (y/n): ").lower()
    if again != "y":
        break
```

## Note
- The pass mark is assumed to be 50.
